```python
#lets convert our float to integers

cols = ['mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area', 'mean_smoothness' ]
cancer_data[cols] = cancer_data[cols].applymap(np.int64)
cancer_data.head(2)
```

```python
STATUSs = LabelEncoder()
UNITSs = LabelEncoder()
Seriess_title_4 = LabelEncoder()
Seriess_title_5 = LabelEncoder()
```

```python
cancer_data['diagnosis'] = STATUSs.fit_transform(cancer_data['diagnosis'])
```

```python
cancer_data.head()
```

```python
def student(diagnosis):
   
    if ( diagnosis== 0 ):
        return ('no')
    else: 
        return ('yes')

cancer_data['i_have_cancer'] = cancer_data['diagnosis'].apply(student)
```
